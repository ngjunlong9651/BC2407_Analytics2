<h2> Data Preprocessing

<p> The Apriori library we are going to use requires our dataset to be in the form of a list of lists, where the whole dataset is a big list and each transaction in the dataset is an inner list within the outer big list. Currently we have data in the form of a pandas dataframe. To convert our pandas dataframe into a list of lists, execute the following script: </p>

```python
## Encoding a wide datatype ## Need to review this codeblock
## Do we even need to encode this given that it is in classical integer representation?
te = TransactionEncoder()
te_ary = te.fit(wide_data).transform(wide_data)
encoded_data = pd.DataFrame(te_ary, columns = te.columns_)
print(encoded_data)
```

```python
## Checking if data is categorical
wide_data
wide_data.dtypes
```

```python
wide_data['milk'] = wide_data['milk'].astype("category")
wide_data['bread'] = wide_data['bread'].astype("category")
wide_data['butter'] = wide_data['butter'].astype("category")
wide_data['beer'] = wide_data['beer'].astype("category")
wide_data['diapers'] = wide_data['diapers'].astype("category")
wide_data
```

```python
encoded_data = pd.get_dummies(wide_data)
encoded_data.dtypes
```

```python
one_hot_encoded_data = pd.get_dummies(wide_data, columns = ["milk", "bread","butter","beer","diapers"])
one_hot_encoded_data.dtypes
```

```python
## running apriori algorithm without encoding to string representation of logical values.
from mlxtend.frequent_patterns import apriori
apriori(one_hot_encoded_data, min_support= 0.4, use_colnames=True)
```
